```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
compilable_tests = {
    "Haiku w/RAG K=1, S=1": 12,
    "Haiku no RAG": 9,
    "Haiku w/RAG K=2, S=2": 34,
}
```

```python
labels = list(compilable_tests.keys())
values = list(compilable_tests.values())

plt.figure(figsize=(10, 6))
sns.barplot(x=labels, y=values, palette='viridis')

plt.ylabel('Cantidad de tests compilables', fontsize=16)

plt.xticks(rotation=45, ha='right', fontsize=12)
plt.grid(axis='y', alpha=0.5)

# Mostrar el gr√°fico
plt.tight_layout()
plt.show()
```

```python
fuzzing_tests = {
    "Haiku w/RAG K=1, S=1": 7,
    "Haiku no RAG": 0.1,
    "Haiku w/RAG K=2, S=2": 16,
}

labels = list(fuzzing_tests.keys())
values = list(fuzzing_tests.values())

plt.figure(figsize=(10, 6))
sns.barplot(x=labels, y=values, palette='viridis')

plt.ylabel('Cantidad de fuzzing tests', fontsize=16)

plt.xticks(rotation=45, ha='right', fontsize=12)
plt.grid(axis='y', alpha=0.5)

# Mostrar el gr√°fico
plt.tight_layout()
plt.show()
```
